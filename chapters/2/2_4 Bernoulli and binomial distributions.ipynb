# 2.4 Bernoulli and binomial distributions

$$
\text{Ber}(y|\theta) = \begin{cases} 
1 - \theta, & \text{if } y = 0 \\
\theta, & \text{if } y = 1
\end{cases}

$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns
sns.set_style("whitegrid")

color = ["#031B88", "#FB7B8E", "#AAB6FB", "#8A2BE2"]
sns.set_palette(color)
```

```python
# success rate
p = 0.3

# length of sample
n = 10000
samples = np.random.binomial(1, p, n)

print("Bernuulli sample of length: ", len(samples))
```

```python

num_successes = np.sum(samples)

success_rate = num_successes / n

print("Count of success in sample:", num_successes)
print("Succes rate:", success_rate)
```

```python
# Probability of success
p = 0.3
# Possible values of the random variable
x = [0, 1]
# Probabilities of the corresponding values
probabilities = [1 - p, p]

plt.bar(x, probabilities)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of the Bernoulli distribution')
plt.xticks(x, ['Failure', 'Success'])
plt.show()
```

$$\text{Bin}(k|N,\theta) = \binom{N}{k} \theta^k (1-\theta)^{n-k}$$

```python
# Number of trials
n = 10
# Probability of success
p = 0.3

# Generate samples from the Binomial distribution
samples = np.random.binomial(n, p, size=10000)

# Plot a histogram of the samples
plt.hist(samples, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.5)
None
```

```python
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf)

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.show()
```

```python
# Plot the Cumulative Distribution Function (CDF) of the Binomial distribution
cdf = binom.cdf(x, n, p)
plt.step(x, cdf, where='post', label='CDF')

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.legend()
plt.show()
```
